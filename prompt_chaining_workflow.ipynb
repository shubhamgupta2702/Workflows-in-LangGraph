```python
from langgraph.graph import StateGraph, START, END
from typing_extensions import TypedDict
from dotenv import load_dotenv
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
```

```python
load_dotenv()
```

```python
llm = HuggingFaceEndpoint(
  repo_id="Qwen/Qwen3-Coder-Next",
  task='text-generation'
)

model = ChatHuggingFace(llm=llm, )
```

```python
class Blog(TypedDict):
  title:str
  outline:str
  content:str
```

```python
def create_outline(state:Blog) -> Blog:
  title = state['title']
  
  prompt = f'Generate an detailed outline on the title:{title}'
  
  outline = model.invoke(prompt).content
  
  state['outline'] = outline
  
  return state
```

```python
def create_blog(state:Blog) -> Blog:
  title = state['title']
  outline = state['outline']
  
  prompt = f'Create a detailed blog on the topic {title} using the detailed outline \n: {outline}'
  
  content = model.invoke(prompt).content
  state['content'] = content
  
  return state
```

```python
graph = StateGraph(Blog)

graph.add_node("create_outline", create_outline)
graph.add_node("create_blog", create_blog)

graph.add_edge(START, "create_outline")
graph.add_edge("create_outline", "create_blog")
graph.add_edge("create_blog", END)

workflow = graph.compile()
```

```python
initial_state = {"title":"Future of GEN AI, ML, DL."}

result = workflow.invoke(initial_state)

print(result)
```

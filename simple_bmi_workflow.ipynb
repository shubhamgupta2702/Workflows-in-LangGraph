```python
from langgraph.graph import StateGraph, START, END
from typing_extensions import TypedDict
```

```python
class State(TypedDict):
  weight: float
  height: float
  bmi: float
  category: str
```

```python
workflow = StateGraph(State)
```

```python
def calculate_bmi(state:State) -> State:
  """Function to calculate bmi"""
  weight = state['weight']
  height = state['height']
  
  bmi = weight /(height*height)
  state['bmi'] = round(bmi, 2)
  
  return state
```

```python
def category_specify(state:State) -> State:
  """Function to put category based on BMI"""
  bmi = state['bmi']
  
  if bmi < 18.5:
    state['category'] = "Underweight"
  elif bmi >=18.5 and bmi <= 24.5:
    state['category'] = "Normal Weight"
  else:
    state['category'] = "Overweight"  
    
  return state
  
```

```python
graph = StateGraph(State)

graph.add_node("calculate_bmi", calculate_bmi)
graph.add_node("category_specification", category_specify)

graph.add_edge(START, 'calculate_bmi')
graph.add_edge('calculate_bmi', 'category_specification')
graph.add_edge('category_specification', END)

workflow = graph.compile()
```

```python
# execute the graph
intial_state = {'weight':60, 'height':1.70}

final_state = workflow.invoke(intial_state)

print(final_state)
```

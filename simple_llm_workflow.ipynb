```python
from langgraph.graph import StateGraph, START, END
from typing_extensions import TypedDict
from dotenv import load_dotenv
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
```

```python
load_dotenv()
```

```python
llm = HuggingFaceEndpoint(
  repo_id="Qwen/Qwen3-Coder-Next",
  task='text-generation'
)

model = ChatHuggingFace(llm=llm, )
```

```python
class State(TypedDict):
  question:str
  answer:str
```

```python
def llm_qa(state:State) -> State:
  """Function for question answering system."""
  question = state['question']
  
  prompt = f'Answer the following question: {question}'
  
  response = model.invoke(prompt)
  
  state['answer'] = response.content
  
  return state
```

```python
graph = StateGraph(State)

graph.add_node("qa_system", llm_qa)

graph.add_edge(START, "qa_system")
graph.add_edge("qa_system", END)

workflow = graph.compile()
```

```python

intial_state = {'question': 'How far is sun from the earth?'}\
  
result = workflow.invoke(intial_state)

print(result['answer'])
```
